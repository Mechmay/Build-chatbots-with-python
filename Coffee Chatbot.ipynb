Whether it’s to get us ready to jump-start our day or to get us through a late-night cram session, many of us need our regular caffeine fix! Ordering coffee is just one example of a process that can be automated with the help of a chatbot.

You’re given the task of creating a Python chatbot that can help cut the wait time of a normal coffee run by taking customers’ orders in advance.

```python
# Define your functions
def coffee_bot():
  print('welcome to the cafe! \nPlease place your orders.')
  order_drink = 'sure'
  drinks = []   
  while order_drink == 'sure':
    size = get_size()
      # print(size)

    drink_type = get_drink_type()

      # print(drink_type)

    hot_iced = hot_or_iced()
    drink = f'{hot_iced} {size} {drink_type}'  

    plas_reuse = plastic_reuse()



    print(f'Alright, that is a {drink} in {plas_reuse}!\n')
    drinks.append(drink)
    while True:
        order_drink = input('Would you like to place another order? (sure/nah) \n> ')
        if order_drink == 'sure' or order_drink == 'nah':
            break
  for i in drinks:
    print('-', i)
  name = input('can I get yout name please? \n> ')  

  print(f'Thanks, {name}! Your drink will be ready shortly.\n')

def get_size():
  res = input('what size drink can I get for you? \n[a] Small \n[b] Medium \n[c] Large \n> ')
  if res == 'a':
    return 'small'
  elif res == 'b':
    return 'medium'
  elif res == 'c':
    return 'large'
  else:
    print_message()
    return get_size()
def print_message():
  print('i\'m sorry, I did not understand your selection. Please enter the corresponding letter for your response.')


def get_drink_type():
  res = input('What type of drink would you like? \n[a] Brewed Coffee \n[b] Mocha \n[c] Latte \n> ')
  if res == 'a':
    return 'brewed coffee'
  elif res == 'b':
    return order_mocha()
  elif res == 'c':
    return order_latte()
  else:
    print_message()
    return get_drink_type()

def order_latte():
  res = input('And what kind of milk for your latte? \n[a] 2% milk \n[b] Non-fat milk \n[c] Soy milk \n> ')

  if res == 'a':
    return '2% milk'
  elif res == 'b':
    return 'non-fat latte'
  elif res == 'c':
    return 'soy latte'
  else:
    print_message()
    return order_latte()

def hot_or_iced():
  res = input('How would you like your drink? \n[a] Hot \n[b]Iced \n> ')
  if res == 'a':
    return 'hot'
  elif res == 'b':
    return 'iced'
  else:
    print_message()
    return hot_or_iced()

def plastic_reuse():
  res = input('Would you like: \n[a] A plastic cup \n[b] your own reusable cup \n> ')
  if res == 'a':
    return 'a plastic cup'
  elif res == 'b':
    return 'your own reusable cup'
  else:
    print_message()
    return plastic_reuse()

def order_mocha():
    while True:
        res = input('Would you like to try our limited-edition peppermint mocha? \n[a] Sure! \n[b] Maybe next time! \n> ')
        if res == 'a':
            return 'peppermint mocha'
        elif res == 'b':
            return 'mocha'
        print_message()

coffee_bot()
```
